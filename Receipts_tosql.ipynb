```python
import sqlite3
import json

# Load the JSON data from a file
with open('receipts.json', 'r') as json_file:
    response = json_file.read()
    response = response.replace('\n', '')
    response = response.replace('}{', '},{')
    response = "[" + response + "]"  # Wrapping in [] to create a valid JSON array
    data = json.loads(response)

# Connect to the SQLite database
conn = sqlite3.connect('data.db')
cursor = conn.cursor()

# Create a table for rewards receipts
cursor.execute('''
    CREATE TABLE IF NOT EXISTS rewards_receipts (
        id TEXT PRIMARY KEY,
        bonusPointsEarned INTEGER,
        bonusPointsEarnedReason TEXT,
        createDate INTEGER,
        dateScanned INTEGER,
        finishedDate INTEGER,
        modifyDate INTEGER,
        pointsAwardedDate INTEGER,
        pointsEarned TEXT,
        purchaseDate INTEGER,
        purchasedItemCount INTEGER,
        rewardsReceiptStatus TEXT,
        totalSpent TEXT,
        userId TEXT
    )
''')

# Iterate through each JSON object and insert data into the table
for entry in data:
    create_date = entry.get('createDate', {}).get('$date', None)
    date_scanned = entry.get('dateScanned', {}).get('$date', None)
    finished_date = entry.get('finishedDate', {}).get('$date', None)
    modify_date = entry.get('modifyDate', {}).get('$date', None)
    points_awarded_date = entry.get('pointsAwardedDate', {}).get('$date', None)
    purchase_date = entry.get('purchaseDate', {}).get('$date', None)

    data_to_insert = (
        entry.get('_id', {}).get('$oid', None),
        entry.get('bonusPointsEarned', None),
        entry.get('bonusPointsEarnedReason', None),
        create_date, date_scanned,
        finished_date, modify_date,
        points_awarded_date, entry.get('pointsEarned', None),
        purchase_date, entry.get('purchasedItemCount', None),
        entry.get('rewardsReceiptStatus', None),
        entry.get('totalSpent', None),
        entry.get('userId', None)
    )

    print(date_scanned)

    insert_query = '''
        INSERT INTO rewards_receipts (
            id, bonusPointsEarned, bonusPointsEarnedReason, createDate,
            dateScanned, finishedDate, modifyDate, pointsAwardedDate,
            pointsEarned, purchaseDate, purchasedItemCount,
            rewardsReceiptStatus, totalSpent, userId
        )
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    '''

    cursor.execute(insert_query, data_to_insert)

# Commit changes and close the connection
conn.commit()
conn.close()
```

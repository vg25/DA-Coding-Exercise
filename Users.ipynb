```python
# Using data types
temporal_columns = df_users.select_dtypes(include=['datetime64'])

# Using heuristics (checking if column name contains "date" or "time")
temporal_columns = [col for col in df_users.columns if 'date' in col.lower() or 'time' in col.lower()]

print("Temporal Columns:")
print(temporal_columns)

categorical_columns = []
for col in df_users.columns:
    if isinstance(df_users[col][0], (str, bool)):
        categorical_columns.append(col)
print ("Categorical Columns are:")
print(categorical_columns)

# Exclude categorical and temporal columns
non_categorical_temporal_columns = [col for col in df_users.columns if col not in categorical_columns + temporal_columns]

# Identify continuous columns
continuous_columns = []
for col in non_categorical_temporal_columns:
    if isinstance(df_users[col][0], (int, float)):
        continuous_columns.append(col)

print("Continuous Columns:")
print(continuous_columns)

# Exclude categorical and temporal columns
non_categorical_temporal_columns = [col for col in df_users.columns if col not in categorical_columns + temporal_columns]

# Identify discrete columns
discrete_columns = []
for col in non_categorical_temporal_columns:
    if isinstance(df_users[col][0], int) and df_users[col].nunique() <= 10:
        discrete_columns.append(col)

print("Discrete Columns:")
print(discrete_columns)
```

```python
#to see the number of values and which are the unique values in the column sign up source
unique_categories = df_users['signUpSource'].unique()

print("Unique values in 'signUpSource' column:")
print(unique_categories)
# Get the number of unique values in the 'category' column
num_unique_categories = df_users['signUpSource'].nunique()

print("Number of unique values in 'signUpSource' column:", num_unique_categories)
```
